```python
import scipy
def scraper(path):
    mat = scipy.io.loadmat(path)
    time = []
    vert_comp = []
    east_comp = []
    north_comp = []

    for i in mat["time"]:
        for j in i:
            time.append(j)
    for i in mat["vert_comp"]:
        for j in i:
            vert_comp.append(j * 9.81)
    for i in mat["east_comp"]:
        for j in i:
            east_comp.append(j * 9.81)
    for i in mat["north_comp"]:
        for j in i:
            north_comp.append(j * 9.81)

    return time, vert_comp, east_comp, north_comp

time, vert_comp, east_comp, north_comp = scraper("/Users/sam/Desktop/Projects/Earthquakes/Assignments/Data/Final_Data/Acadia_RSN2320.mat")
print(time, vert_comp, east_comp, north_comp)
```
